### Creating a Decision Tree

Consider the following two-dimensional data, which has one of four class labels (see the following figure):

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=300, centers=4,
                  random_state=0, cluster_std=1.0)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='rainbow');
```

# In the code/graph below
* 'n_samples=5000': This parameter determines the total number of data points generated. Previously, it was generating 300 data points, but now it creates a dataset with 5000 points.
* 'centers=10': The centers parameter defines the number of clusters or centers around which the data points are generated. Initially, it generated data around 4 centers, but now it creates data around 10 centers.

```python
X, y = make_blobs(n_samples=5000, centers=10,
                  random_state=0, cluster_std=1.0)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='rainbow');
```

# In the code/graph below
* 'random_state': This parameter determines the random seed used by the random number generator. When I changed it from 0 to 2, it altered the random seed. Different random seeds will produce different random data, even if the other parameters are the same. Therefore, changing the 'random_state' will result in a different arrangement of points in the dataset.
* 'cluster_std': This parameter sets the standard deviation of the clusters. When increased from 1 to 5, it widens the spread of points within each cluster. A higher standard deviation means the points in each cluster will be more spread out from the cluster's center, making the clusters less compact and more dispersed.

```python
X, y = make_blobs(n_samples=5000, centers=10,
                  random_state=2, cluster_std=5.0)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='rainbow');
```

# In the code/graph below
1. Changed X[:, 0] to X[:, 1]:
    * Changing it to X[:, 1] means that using the values of the second feature/column for the x-coordinates in the scatter plot. This switch implies plotting the data points using the values from a different feature or dimension.
2. Changed X[:, 1] to X[:, 0]:
    * Now, the plot uses the first feature's values as y-coordinates, effectively swapping the axes for visualization.
3. Changed s=50 to s=100:
    * Increasing it from 50 to 100 increases the size of the points, making them visually larger in the plot.      

```python
X, y = make_blobs(n_samples=300, centers=4,
                  random_state=0, cluster_std=1.0)
plt.scatter(X[:, 1], X[:, 0], c=y, s=100, cmap='rainbow');
```
